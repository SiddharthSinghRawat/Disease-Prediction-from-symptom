## Transforming & Loading the Data

### Loading the Dataset File generated after preprocessing in excel

```python
disease_list = []

def return_list(disease):
    disease_list = []
    match = disease.replace('^','_').split('_') # using _ as common splitting delimeter
    ctr = 1
    for group in match:
        if ctr%2==0:
            disease_list.append(group) # refer the data format
        ctr = ctr + 1

    return disease_list

with open("raw_data_2.csv") as csvfile:
    reader = csv.reader(csvfile)
    disease=""
    weight = 0
    disease_list = []
    dict_wt = {}
    dict_=defaultdict(list)
    
    for row in reader:

        if row[0]!="\xc2\0xa0" and row[0]!="": # for handling file encoding errors
          # saving disease and frequency
            disease = row[0]
            disease_list = return_list(disease)
            weight = row[1]

        if row[2]!="\xc2\0xa0" and row[2]!="":
            symptom_list = return_list(row[2])

            for d in disease_list:
                for s in symptom_list:
                    dict_[d].append(s) # adding all symptoms
                dict_wt[d] = weight
```

### Reformatting the data

```python
# saving cleaned data
with open("dataset_clean.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    for key,values in dict_.items():
        for v in values:
            #key = str.encode(key)
            key = str.encode(key).decode('utf-8')
            #.strip()
            #v = v.encode('utf-8').strip()
            #v = str.encode(v)
            writer.writerow([key,v,dict_wt[key]])
```

```python
import pandas as pd

# Assuming 'df' is your dataset
# Example structure of df (prognosis column for disease names, other columns for symptoms):
# df = pd.read_csv('your_dataset.csv')

def get_symptoms_for_disease(df, disease):
    
    # Filter the DataFrame for the specified disease
    disease_df = df[df['prognosis'] == disease]
    
    if disease_df.empty:
        return f"No data found for disease: {disease}"
    
    # Sum the symptom columns to get a count of occurrences
    symptom_counts = disease_df.drop(columns=['prognosis']).sum()
    
    # Extract symptoms that are present (count > 0)
    symptoms = symptom_counts[symptom_counts > 0].index.tolist()
    
    return symptoms

# Example usage to get symptoms for 'common_cold'
disease_name = ''
symptoms = get_symptoms_for_disease(df, disease_name)

print(f"Symptoms associated with {disease_name}:")
print(symptoms)
```

```python
from ipywidgets import SelectMultiple, Output
import numpy as np

# Assuming you have a DataFrame named 'df' containing your dataset
# Get all column names as symptoms, excluding "prognosis"
all_symptoms = df.columns[df.columns != 'prognosis'].tolist()

# Create a widget for symptom selection
symptom_select = SelectMultiple(
    options=all_symptoms,
    description='Select Symptoms:',
    disabled=False
)

# Display the widget
display(symptom_select)

output = Output()

# Assuming 'cols' represents feature names used during training
cols = all_symptoms

def on_change(change):
    with output:
        output.clear_output()
        selected_symptoms = list(change['new'])
        if len(selected_symptoms) > 0:
            # Initialize ipt with the correct number of features
            ipt = [0] * len(cols)
            for s in selected_symptoms:
                if s in cols:
                    # Use try-except block to handle cases where a selected symptom is not in cols
                    try:
                        ipt[cols.index(s)] = 1
                    except ValueError:
                        pass
            ipt = np.array([ipt])

            # Ensure consistency in the number of features
            if ipt.shape[1] == len(cols):
                # Print input and prediction results
                print("Selected Symptoms: ", selected_symptoms)
                print("Input for Prediction: ", ipt)
                print("Predicted Disease: ", dt.predict(ipt))
                print("Prediction Probabilities: ", dt.predict_proba(ipt))
            else:
                print("Inconsistent number of features.")
        else:
            print("No symptoms selected.")

# Attach the change handler to the widget
symptom_select.observe(on_change, names='value')

# Display the output
display(output)
```
